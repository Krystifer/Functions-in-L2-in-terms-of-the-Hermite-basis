```python
from matplotlib import pyplot as plt
from numpy import math as npm
import numpy as np
from scipy.integrate import quad
```

```python
def Hermite_polynomial(x,n):  # Defines the Hermite polynomial of order n.
    
    def polynomial_Coefficient(x,n,j):   # Coefficients of the powers of the Hermite polynomial of order n
        return ((npm.factorial(n)*((-1)**(j))*(2**(n-2*j))*(x**(n-2*j)))/((npm.factorial(j))*(npm.factorial(n-2*j))))

    Hermite_polynomial = 0          
    if n%2==0:
        for j in range(int(n/2.)+1): # The '+1' is require due to the range() function doesn't take the upper bound of the interval
            Hermite_polynomial+= polynomial_Coefficient(x,n,j)
    else:
        for j in range(int((n-1)/2.)+1):
            Hermite_polynomial += polynomial_Coefficient(x,n,j)
    return Hermite_polynomial
```

```python
def Hermite_function(x,n): # Hermite function of order n
    return (np.exp(-(1./2)*x**2)*Hermite_polynomial(x,n))/(((2**n)*npm.factorial(n)*(np.pi)**(1./2))**(1./2))
```

```python
def scalar_Approximation(x,n,f): # Returns the coefficient n of the linear combination of f in terms of the Hermite functions
    producto_Interno= quad(lambda x: f(x)*Hermite_function(x,n), -np.inf, np.inf)[0] 
    return producto_Interno
```

```python
def f_Series(x,m,f): # You can change the amount of addends that are calculated (m), to obtain better results
    f_Series = ((np.exp(-(x**2)/2.))/(np.pi**(1./4)))*quad(lambda x: f(x)*np.exp(-(x**2)/2.)/(np.pi**(1./4)), -np.inf, np.inf)[0] #t√©rmino alfa_0*H_0
    for n in range(1, int(m)):
        f_Series = f_Series+ scalar_Approximation(x,n,f)*Hermite_function(x,n)
    return f_Series
```

## Let's define several functions and find their approximations.

```python
#Define f, which is the function that we want to approximate by Hermite functions.
def f(x):
    return np.piecewise(x, [np.logical_and(-1<x,x<1), ], [1, 0])
def g(x):
    return x
def h(x):
    return np.sin(x)
def j(x):
    return np.piecewise(x, [np.logical_and(-1<x,x<1), np.logical_and(-5<x,x<-1), np.logical_and(-10<x,x<-5) ], [1, -10,15])
```

```python
x= np.linspace(-5,5,1000)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.xlim([-5,5])
plt.ylim([-0.2,1.2])
plt.grid(linewidth=.5, linestyle='-')

plt.plot(x, np.piecewise(x, [np.logical_and(-1<x,x<1), ], [1, 0]), color='black',label='f(x)')
plt.plot(x,f_Series(x,2,f), color="steelblue", label='n=2')
plt.plot(x,f_Series(x,30,f), color="green", linewidth=1.5, label='n=30')
plt.plot(x,f_Series(x,70,f), color="darkmagenta", linewidth=1.5, label='n=70')

plt.legend()
```

```python
x= np.linspace(-5,5,1000)
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid(linewidth=.5, linestyle='-')

plt.plot(x, g(x))
plt.plot(x,f_Series(x,40,g), color="darkmagenta", label='n=40')

plt.legend()
```

```python
x= np.linspace(-20,20,1000)
plt.xlabel("x")
plt.ylabel("h(x)")
plt.grid(linewidth=.5, linestyle='-')

plt.plot(x, h(x))
plt.plot(x,f_Series(x,20,h), color="darkred", label='n=20')

plt.legend()
```
